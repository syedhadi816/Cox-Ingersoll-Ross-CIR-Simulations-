# CIR Simulation

```python
import math
import numpy as np

def cir(r0, Y, theta, sigma, T=1., N=10,seed=777):
    np.random.seed(seed)
    dt = T/float(N)    
    rates = [r0]
    d_r = []
    for i in range(N):
        dr = Y*(theta-rates[-1])*dt + sigma*math.sqrt(np.abs(rates[-1]))*np.random.normal()
        rates.append(rates[-1] + dr)
        d_r.append(dr)
    return range(N+1), rates, d_r
```

```python
x, y, dr = cir(0.01 , 0.20, 0.01, 0.012, 10., 200)
```

```python
import matplotlib.pyplot as plt
plt.plot(x,y)
plt.show()
```

# House Price Monte Carlo Simulation

```python
def hprice(a0, H, sigma_h, rho , T=1., N=10,seed=777):
    np.random.seed(seed)
    dt = T/float(N)    
    h_price = [a0]
    d_H = []
    for i in range(N):
        dH = a0*(h_price[-1])*dt + sigma_h*rho*(h_price[-1])*np.random.normal() + sigma_h*(math.sqrt(1-rho**2))*(h_price[-1])*np.random.normal()
        h_price.append(h_price[-1] + dH)
        d_H.append(dH)
    return range(N+1), h_price, d_H
```

```python
x, y, dH = hprice(0.01 , 1 , 0.012, 0.4 ,10., 200)
```

```python
import matplotlib.pyplot as plt
plt.plot(x,y)
plt.show()
```
